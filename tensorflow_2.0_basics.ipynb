```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
observation = 1000
xs = np.random.uniform(high=10,low=-10,size=(observation,1))
zs = np.random.uniform(10,-10,size=(observation,1))
generated_inputs = np.column_stack((xs,zs))
noise = np.random.uniform(1,-1,size=(observation,1))
generated_targets = 2 * xs - 3 * zs + 5 + noise
np.savez('TF_intro', input = generated_inputs, target = generated_targets)
```

```python
training_data = np.load('TF_intro.npz')
input_size = 2
output_size = 1
model = tf.keras.Sequential([tf.keras.layers.Dense(output_size,
                                                  kernel_initializer = tf.random_uniform_initializer(minval=-0.1,maxval=0.1),
                                                  bias_initializer = tf.random_uniform_initializer(minval=-0.1,maxval=0.1)
                                                  )
                            ])
custom_optimizer = tf.keras.optimizers.SGD(learning_rate=0.02)
model.compile(optimizer="SGD",loss='mean_squared_error')
model.fit(training_data['input'],training_data['target'],epochs=100,verbose=2)
```
